```python
# === Estrutura dos dados do MNIST usando pandas e numpy ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
```

```python
# === Passo 1: Carregar o conjunto de dados MNIST ===
# O dataset vem separado em treino e teste
(X_treino, y_treino), (X_teste, y_teste) = mnist.load_data()
```

```python
# === Passo 2: Ver informações básicas ===
print("Formato de X_treino:", X_treino.shape)
print("Formato de y_treino:", y_treino.shape)
print("Formato de X_teste:", X_teste.shape)
print("Formato de y_teste:", y_teste.shape)
```

```python
# === Passo 3: Converter para DataFrame (estrutura tabular) ===
# Cada imagem é convertida em uma linha de 784 colunas (28x28 pixels achatados)
num_imagens, altura, largura = X_treino.shape
X_treino_flat = X_treino.reshape(num_imagens, altura * largura)

# Cria DataFrame pandas para inspeção
df_mnist = pd.DataFrame(X_treino_flat)
df_mnist["label"] = y_treino  # adiciona rótulo como última coluna
```

```python
# === Passo 4: Mostrar estrutura dos dados ===
print("\nExemplo das primeiras linhas do MNIST em formato tabular:")
print(df_mnist.head())

print("\nInformações sobre o DataFrame:")
print(df_mnist.info())
```
