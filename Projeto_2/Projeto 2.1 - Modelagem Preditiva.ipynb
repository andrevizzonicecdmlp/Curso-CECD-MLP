```python
# Importando bibliotecas
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
```

```python
# Carregando o dataset MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# Normalizando os dados (0–255 → 0–1)
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# Convertendo rótulos em one-hot encoding
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
# Construindo a rede neural
model = Sequential([
    Flatten(input_shape=(28, 28)),  # Transformando imagem 2D em vetor
    Dense(128, activation='relu'),  # Camada escondida com 128 neurônios
    Dense(10, activation='softmax')  # Saída com 10 classes (0 a 9)
])
```

```python
# Compilando o modelo
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Treinando a rede
model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.1)
```

```python
# Avaliando no teste
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f'Acurácia no teste: {test_acc:.2%}')
```
